In this notebook, I collect the business/first class passenger ratio on a given origin-destination pair (on a single flight) for a given quarter from the DB1B Coupon data collected for 2016 and 2017 for the origin and destination airports of interest.

```python
import pickle
import pandas as pd
import numpy as np
```

Read in the saved DB1B Coupon data

```python
with open('db1b_coup.pkl', 'rb') as file1:
    db1b_coup = pickle.load(file1)
```

```python
db1b_coup.head()
```

Compute number of business passengers (Coded in `FareClass` as C or D) and all passengers for a given Year, Quarter and O-D pair 4-tuple. Then get the ratio of business passengers to all passengers and save the file.

```python
bus_pass = db1b_coup[db1b_coup['FareClass'].isin(['C', 'D'])].groupby(['Year', 'Quarter', 'Origin', 'Dest']).sum()[['Passengers']]
```

```python
all_pass = db1b_coup.groupby(['Year', 'Quarter', 'Origin', 'Dest']).sum()[['Passengers']]
```

```python
bus_pass_ratio = (bus_pass/all_pass).fillna(0).reset_index()
bus_pass_ratio.rename({'Passengers': 'BusPassRatio'}, axis = 1, inplace=True)
```

```python
bus_pass_ratio
```

```python
with open('bus_pass_ratio.pkl', 'wb') as file1:
    pickle.dump(bus_pass_ratio, file1)
```
