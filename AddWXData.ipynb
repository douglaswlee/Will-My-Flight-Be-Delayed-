Read in the weather stations in [NOAA's Global Historical Climate Network Daily (GHCND)](https://www.ncdc.noaa.gov/ghcn-daily-description) dataset and do some formatting.

```python
with open('WXData/ghcnd-stations.txt', 'rb') as f:
    station_data = [lines.strip() for lines in f]
```

Put each row (station data) into its own list.

```python
station_data = [re.split('\s+', x) for x in station_data]
```

Get just the id, latitude and longitude of each station (the first three elements of each row list).

```python
station_data = [x[:3] for x in station_data]
```

```python
station_data
```

Convert everything into a dataframe.

```python
import pandas as pd
station_data = pd.DataFrame(station_data, columns=['Station', 'Latitude', 'Longitude'])
station_data.head()
```

Get the latitutde and longitudes of all airports we are evaluating from saved file.

```python
import pickle
with open('lat_longs.pkl', 'rb') as pickz:
    lat_longs = pickle.load(pickz)
```

```python
import numpy as np
import geopy.distance
```

Combine latitude and longitude into a tuple in the dataframe for weather station data. 

```python
station_data['LatLong'] = list(zip(station_data['Latitude'].astype('float'), station_data['Longitude'].astype('float')))
```

Read in and combine GHCND weather data for all stations from both 2016 and 2017

```python
ghcnd16 = pd.read_csv('WXData/2016.csv', header=None)
ghcnd17 = pd.read_csv('WXData/2017.csv', header=None)
ghcnd18 = pd.read_csv('WXData/2018.csv', header=None)
```

```python
ghcnd_all = pd.concat([ghcnd16, ghcnd17], axis = 0)
ghcnd_all = pd.concat([ghcnd_all, ghcnd18], axis = 0)
```

Keep only the columns which give the station, date, observation (weather feature being observed -- i.e., average temperature, precipiation for a given day), and value of that observation. Then rename all columns.

```python
ghcnd_all = ghcnd_all.iloc[:, :4]
ghcnd_all.rename({0: 'Station', 1:'Date', 2: 'Obs', 3: 'Value'}, axis=1, inplace=True)
ghcnd_all.head()
```

Code to ensure that weather stations are only associated with those that have records in the GHCND dataset.

```python
stations_ghcnd = list(ghcnd_all['Station'].unique())
```

```python
station_data = station_data[station_data['Station'].isin(stations_ghcnd)]
```

Create a function that makes use of `geopy` to use the [Vincenty formulae](https://en.wikipedia.org/wiki/Vincenty%27s_formulae) to calculate distances between weather stations and airport and find the nearest station to each airport under evaluation.

```python
def find_nearest_station_to_airport(airport_code):
    dist_from = list(station_data['LatLong'].apply(lambda x: geopy.distance.vincenty(x, lat_longs[airport_code])))
    min_dist_idx = np.argmin(dist_from)
    return station_data.iloc[min_dist_idx]['Station']
```

Get our list of airports under evaluation from file and populate a dictionary for the nearest weather station to each airport.

```python
with open('airports_for_model.pkl', 'rb') as p:
    airports = pickle.load(p)
```

```python
nearest_station = {}
for j in airports:
    nearest_station[j] = find_nearest_station_to_airport(j)
```

```python
nearest_station
```

Keep only the closest station to each airport being evaluated.

```python
ghcnd_keep = ghcnd_all[ghcnd_all['Station'].isin(nearest_station.values())]
```

```python
ghcnd_keep.shape
```

```python
ghcnd_keep = ghcnd_keep.reset_index()
```

Now, keep only the following observations from each of the stations for each day:
* Average daily temperature
* Total daily precipitation
* Average daily wind speed

```python
obs_keep = ['TAVG', 'PRCP', 'AWND']
ghcnd_keep = ghcnd_keep[ghcnd_keep['Obs'].isin(obs_keep)].reset_index()
```

Match airport to station for the GHCND data.

```python
ghcnd_keep['Airport'] = [k for station in ghcnd_keep['Station'] for k,v in nearest_station.items() if v==station]
```

```python
ghcnd_keep['Airport'].nunique()
```

```python
ghcnd_keep.head()
```

Keep only the columns from above needed to merge with the AOTP data.

```python
ghcnd_keep = ghcnd_keep[['Date', 'Obs', 'Value', 'Airport']]
```

```python
ghcnd_keep.shape
```

Reformat dates to match format of AOTP data.

```python
ghcnd_keep.rename({'Date': 'dt'}, axis=1, inplace=True)
ghcnd_keep['Date'] = [str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[-2:] for x in ghcnd_keep['dt']]
```

```python
ghcnd_keep.drop('dt', axis=1, inplace=True)
```

```python
ghcnd_keep.head()
```

Pull out, respectively, just the average temperature, total precipitation and average wind speed for each into their own dataframes (currently in long format). Then collect each into a dictionary keyed by an airport-date tuple.

```python
atemp = ghcnd_keep[ghcnd_keep['Obs'] == 'TAVG']
precip = ghcnd_keep[ghcnd_keep['Obs'] == 'PRCP']
awind = ghcnd_keep[ghcnd_keep['Obs'] == 'AWND']
```

```python
temp_dict = atemp[['Airport', 'Date', 'Value']].set_index(['Airport', 'Date']).to_dict()['Value']
```

```python
precip_dict = precip[['Airport', 'Date', 'Value']].set_index(['Airport', 'Date']).to_dict()['Value']
```

```python
wind_dict = awind[['Airport', 'Date', 'Value']].set_index(['Airport', 'Date']).to_dict()['Value']
```

Read in the AOTP data to merge with the GHCND data. Also save the GHCND data we are using to have access to it later.

```python
with open('aotp_nyc_new.pkl', 'rb') as picks:
    aotp_nyc_data = pickle.load(picks)
```

```python
with open('ghcnd.pkl', 'wb') as p2:
    pickle.dump(ghcnd_keep, p2)
```

Again, make dataframes (now in wide format) for average temperature, total precipitation and average wind speed for each airport-date combination. 

```python
temps = pd.DataFrame.from_dict(temp_dict, orient='index')
temps = temps.reset_index()
temps[['Airport', 'Date']] = temps['index'].apply(pd.Series)
temps.rename({0: 'ATemp'}, axis=1, inplace=True)
temps.drop('index', axis=1, inplace=True)
temps.head()
```

```python
precip = pd.DataFrame.from_dict(precip_dict, orient='index')
precip = precip.reset_index()
precip[['Airport', 'Date']] = precip['index'].apply(pd.Series)
precip.rename({0: 'Precip'}, axis=1, inplace=True)
precip.drop('index', axis=1, inplace=True)
```

```python
precip.head()
```

```python
AWind = pd.DataFrame.from_dict(wind_dict, orient='index')
AWind = AWind.reset_index()
AWind[['Airport', 'Date']] = AWind['index'].apply(pd.Series)
AWind.rename({0: 'AWind'}, axis=1, inplace=True)
AWind.drop('index', axis=1, inplace=True)
```

```python
AWind.head()
```

Separate each of the weather measures by origin and destination airports. Then rename columns to match those in the AOTP data and to differntiate between origin and destination.

```python
tempsOrig = temps[temps['Airport'].isin(['EWR', 'JFK', 'LGA'])]
tempsDest = temps[~temps['Airport'].isin(['EWR', 'JFK', 'LGA'])]
```

```python
precipOrig = precip[precip['Airport'].isin(['EWR', 'JFK', 'LGA'])]
precipDest = precip[~precip['Airport'].isin(['EWR', 'JFK', 'LGA'])]
```

```python
windOrig = AWind[AWind['Airport'].isin(['EWR', 'JFK', 'LGA'])]
windDest = AWind[~AWind['Airport'].isin(['EWR', 'JFK', 'LGA'])]
```

```python
tempsOrig.rename({'Airport':'Origin', 'Date':'FlightDate'}, axis=1,  inplace=True)
tempsDest.rename({'Airport':'Dest', 'Date': 'FlightDate', 'ATemp': 'ATempDest'}, axis=1,  inplace=True)
```

```python
precipOrig.rename({'Airport':'Origin', 'Date': 'FlightDate'}, axis=1,  inplace=True)
precipDest.rename({'Airport': 'Dest', 'Date': 'FlightDate', 'Precip': 'PrecipDest'}, axis=1, inplace=True)
```

```python
windOrig.rename({'Airport':'Origin', 'Date': 'FlightDate'}, axis=1,  inplace=True)
windDest.rename({'Airport': 'Dest', 'Date': 'FlightDate', 'AWind': 'AWindDest'}, axis=1, inplace=True)
```

Now, merge everything.

```python
tmp = aotp_nyc_data.merge(tempsOrig, on=['FlightDate', 'Origin'], how='left')
```

```python
tmp = tmp.merge(precipOrig, on = ['FlightDate', 'Origin'], how='left')
```

```python
tmp = tmp.merge(windOrig, on=['FlightDate', 'Origin'], how='left')
```

```python
tmp = tmp.merge(tempsDest, on=['FlightDate', 'Dest'], how='left')
```

```python
tmp = tmp.merge(precipDest, on=['FlightDate', 'Dest'], how='left')
```

```python
tmp = tmp.merge(windDest, on=['FlightDate', 'Dest'], how='left')
```
